## 3.2 Platform Sales Analysis Over Time

Now let's analyze how sales vary across platforms and years:

```python
# Calculate total sales by platform and year
platform_sales_year = games.groupby(['platform','year_of_release'])['total_sales'].sum()
platform_sales_year = platform_sales_year.sort_values(ascending=False)
display(platform_sales_year)
```

```python
# Create a heatmap of platform sales over time
sales_pivot = games.pivot_table(
    values='total_sales',
    index='platform',
    columns='year_of_release',
    aggfunc='sum',
    fill_value=0
)

display(sales_pivot)
```

```python
# Create the heatmap
plt.figure(figsize=(15, 10))
sns.heatmap(sales_pivot, 
            cmap='YlOrRd',  # Color scheme (yellow to red)
            annot=False,    # Don't show numbers in each cell (too crowded)
            fmt='.1f',      # Format for numbers if you want them
            cbar_kws={'label': 'Total Sales (millions)'})

plt.title('Platform Sales Over Time Heatmap')
plt.xlabel('Year of Release')
plt.ylabel('Platform')
plt.xticks(rotation=45)  # Rotate year labels for better readability
plt.tight_layout()
plt.show()

#This is the first time I've had to create a heatmap. I've used the help of Dot for this and have taken notes on how this works
#as it wasn't shown or taught in the lession/sprint.
```

```python
# Identify platforms with declining sales
# Calculate year-over-year percentage change for each platform
yoy_changes = sales_pivot.pct_change(axis=1) * 100  # axis=1 for columns (years)

display(yoy_changes)
```

```python
# Find platforms with the most significant negative changes
# Look for platforms with large negative percentages
print("Platforms with significant year-over-year declines:")
print("\nMost negative changes by platform:")

# Find the minimum (most negative) change for each platform
min_changes = yoy_changes.min(axis=1).sort_values()
print(min_changes.head(10))

# Find platforms that went from having sales to zero sales
print("\nPlatforms that completely disappeared from market:")

# Look for platforms where sales went to 0 and stayed there
for platform in sales_pivot.index:
    platform_data = sales_pivot.loc[platform]
    # Find if platform had sales then went to 0
    if platform_data.max() > 0:  # Had sales at some point
        last_sales_year = platform_data[platform_data > 0].index.max()
        if last_sales_year < sales_pivot.columns.max():  # Didn't sell until the end
            print(f"{platform}: Last sales in {last_sales_year}")

# Count consecutive years of decline for each platform
print("\nPlatforms with consistent declining trends:")

for platform in yoy_changes.index:
    negative_changes = yoy_changes.loc[platform] < -10  # More than 10% decline
    consecutive_declines = 0
    max_consecutive = 0
    
    for change in negative_changes:
        if change:
            consecutive_declines += 1
            max_consecutive = max(max_consecutive, consecutive_declines)
        else:
            consecutive_declines = 0
    
    if max_consecutive >= 2:  # At least 2 consecutive years of decline
        print(f"{platform}: {max_consecutive} consecutive years of >10% decline")
```

# Step 5 : Hypothesis Tests

‚ÄîAverage user ratings of the Xbox One and PC platforms are the same.

‚ÄîAverage user ratings for the Action and Sports genres are different.

Set the *alpha* threshold value yourself.

Explain:

‚ÄîHow you formulated the null and alternative hypotheses

‚ÄîWhat criteria you used to test the hypotheses~~,~~ and why


```python
print(f"Null Hypothesis: Average user ratings of the Xbox One and PC platforms are the same.")
print(f"Alternative Hypothesis: Average user ratings of the Xbox One and PC platforms are different.")

alpha = 0.05

xOne_sample = games_relevant[games_relevant['platform'] == 'XOne'].dropna()['user_score']
pc_sample = games_relevant[games_relevant['platform'] == 'PC'].dropna()['user_score']
results = st.ttest_ind(xOne_sample, pc_sample, equal_var=False)
print('p-value:', results.pvalue)

if(results.pvalue < alpha):
    print('We reject the null hypothesis')
else:
    print('We cannot reject the null hypothesis')

# The platforms have same or similar average user ratings.
```
